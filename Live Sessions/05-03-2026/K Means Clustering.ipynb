```python
'''
Unsupervised Learning:
Imagine you enter a supermarket with thousand customers
company does not know: who like electronics, who buys groceries, etc

IF we group similar customers automatically, we can discover the pattern

What is unsupervised learning ?

The type of machine learning: no labelled output is given, algorithm finds the HIDDEN PATTERN

Types:
1. Clustering - group similar data points
eg. customer segmentation, document grouping, image grouping

2.Associations - which will find the relationship
eg. market basket analysis(if someone buys bread they are likely to buy butter or jam)
'''
```

```python
'''
K-MEANS CLUSTERING:
Groups the data into k clusters.
Where K = No of Groups

CUSTOMER DATA:
AGE    INCOME
22     30k
23     28k
45     90k
46     88k

Cluster 1 -> Young customers
Cluster 2 -> High Income Customers


How does K-Means Algorithm work:

Step 1: Choose K, eg K=2

Step 2: Initialise Centroids
Algorithm Randomly Selects 2 Center Points

Ex: C1,C2

Step 3: Assign Points to nearest centroid
Each data point goes to closet centroid
Euclidean Distance Formula

EX:
Point(2,3)
CENTROID 1=(1,1)
CENTROID 2=(5,6)

Step 4: Update the centroid: average of all points in cluster
CLuster Points: (2,3)(3,4)(4,5)
New Centroid = (2+3+4)/3,(3+4+5)/3
=(3,4)

Step 5: Repeat until STABLE

ASSIGN -> UPDATE -> ASSIGN -> UPDATE
'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#create datasets
data = {
    "Age": [22,23,24,45,46,47,30,31,32],
    "Income":[25,28,26,90,88,95,60,65,62]
}
df = pd.DataFrame(data)
print(df)
```

```python
#visulaise the data
plt.scatter(df["Age"],df["Income"])
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Data visualise")
plt.show()
```

```python
#apply kmeans
kmeans = KMeans(n_clusters=1)
kmeans.fit(df)
```

```python
label = kmeans.labels_
print(label)
```

```python
'''
cluster 0 -> young
cluster 1 -> high income
cluster 2 ->
'''
```

```python
#visulaise the data
plt.scatter(df["Age"],df["Income"],c=label)
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Data visualise")
plt.show()
```

```python
#show centroids
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df['Age'],df['Income'],c=label)
plt.scatter(centroids[:,0],centroids[:,1],marker='X',s=200)
plt.show()
```
