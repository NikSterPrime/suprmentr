```python
import time
import numpy as np

python_list = list(range(1_000_000))
numpy_array = np.arange(1_000_000)
```

```python
start = time.time()
new_list = [x+5 for x in python_list]
end = time.time()
print("Python List addition time: ",end-start)

start = time.time()
new_array = numpy_array + 5
end = time.time()
print("NumPy Array addition time: ",end-start)
```

```python
start = time.time()
new_list = [x*2 for x in python_list]
end = time.time()
print("Python List multiplication time: ",end-start)

start = time.time()
new_array = numpy_array * 5
end = time.time()
print("NumPy Array multiplication time: ",end-start)
```

```python
start = time.time()
total_list = sum(python_list)
end = time.time()
print("Python List multiplication time: ",end-start)

start = time.time()
total_array = np.sum(numpy_array)
end = time.time()
print("NumPy Array multiplication time: ",end-start)
```
