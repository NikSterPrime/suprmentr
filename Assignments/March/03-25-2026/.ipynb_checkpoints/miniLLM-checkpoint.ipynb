```python
import random
from gensim.models import Word2Vec
import numpy as np
```

```python
sentences = [
    "college life is fun and exciting",
    "students attend lectures daily",
    "exams are stressful for students",
    "friends make college life memorable",
    "assignments and projects are important",
    "library is a quiet place to study",
    "group study helps in understanding concepts",
    "college festivals are enjoyable",
    "students participate in sports and events",
    "hostel life teaches independence",
    "late night study is common during exams",
    "teachers guide students for better learning",
    "practical knowledge is important in college",
    "students prepare for placements",
    "internships help students gain experience"
]
```

```python
tokenized = [s.lower().split() for s in sentences]
```

```python
model = Word2Vec(sentences=tokenized, vector_size=50, window=3, min_count=1)
```

```python
def predict_next(context_words):
    vectors = []
    for word in context_words:
        if word in model.wv:
            vectors.append(model.wv[word])
    if not vectors:
        return random.choice(list(model.vw.index_to_key))
    avg_vector = np.mean(vectors, axis=0)
    similar = model.wv.similar_by_vector(avg_vector, topn=5)
    return random.choice(similar)[0]
```

```python
def generate_sentence(start_text, length=10):
    words = start_text.lower().split()
    
    while len(words) < length:
        next_word = predict_next(words[-2:])  # use last 2 words as context
        words.append(next_word)
    
    return " ".join(words)
```

```python
user_input = input("Enter a starting sentence: ")
output = generate_sentence(user_input,len(user_input))
print("\nGenerated Sentence:")
print(output)
```
