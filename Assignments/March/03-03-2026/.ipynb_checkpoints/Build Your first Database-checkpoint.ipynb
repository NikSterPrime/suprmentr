```python
# Assignment: Build Your First Dataset

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# ---------------------------------------------------
# Step 1: Create Dataset
# ---------------------------------------------------

data = {
    "Study_Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks": [35, 40, 50, 55, 65, 70, 80, 90]
}

df = pd.DataFrame(data)

print("Dataset:")
display(df)


# ---------------------------------------------------
# Step 2: Identify Features and Labels
# ---------------------------------------------------

# Feature (Input)
X = df[["Study_Hours"]]

# Label (Output)
y = df["Marks"]

print("\nFeature (X): Study Hours")
print("Label (y): Marks")


# ---------------------------------------------------
# Step 3: Train Linear Regression Model
# ---------------------------------------------------

model = LinearRegression()
model.fit(X, y)


# ---------------------------------------------------
# Step 4: Predict Relationship
# ---------------------------------------------------

predicted_marks = model.predict([[5]])
print("\nPredicted marks for studying 5 hours:", predicted_marks[0])


# ---------------------------------------------------
# Step 5: Visualize Relationship
# ---------------------------------------------------

plt.scatter(df["Study_Hours"], df["Marks"], label="Actual Data")
plt.plot(df["Study_Hours"], model.predict(X), label="Regression Line")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.legend()
plt.show()
```
