```python
import pandas as pd

# Prompt comparison
data = {
    "Task": ["Resume", "Business Idea", "Study Plan"],
    
    "Weak Prompt": [
        "Write resume",
        "Give business idea",
        "Make study plan"
    ],
    
    "Strong Prompt": [
        "Write a professional resume for a computer science student with skills in Python and ML",
        "Suggest a profitable AI-based business idea for students with low investment",
        "Create a 2-week study plan for machine learning with daily tasks"
    ]
}

df = pd.DataFrame(data)

print("Prompt Comparison:\n")
print(df)
```

## 🧠 Assignment (30/03/2026): Prompt Engineer

### 📌 Objective  
The objective of this assignment is to understand how the **quality of prompts affects AI-generated responses**. By comparing weak and strong prompts, we can observe how better instructions lead to more accurate and useful outputs.

---

### ⚙️ Methodology  

1. **Create Tasks**  
   Selected common tasks such as:
   - Resume writing  
   - Business idea generation  
   - Study planning  

2. **Write Weak Prompts**  
   Weak prompts are vague and lack detail.  
   Example:
   - "Write resume"  
   - "Give business idea"  

3. **Write Strong Prompts**  
   Strong prompts include clear instructions, context, and constraints.  
   Example:
   - "Write a professional resume for a computer science student with skills in Python and Machine Learning"  
   - "Suggest a profitable AI-based business idea for students with low investment"  

4. **Compare Outputs**  
   The outputs generated from strong prompts are more:
   - Relevant  
   - Detailed  
   - Structured  

---

### 📊 Observations  

| Aspect | Weak Prompt | Strong Prompt |
|--------|------------|--------------|
| Clarity | Low | High |
| Detail | Minimal | Detailed |
| Output Quality | Generic | Specific & useful |
| Context | Missing | Clearly defined |

---

### 🧠 Key Insight  
AI models depend heavily on the **input prompt quality**. A well-structured prompt guides the model to produce more meaningful and accurate results.
