```python
from gensim.models import Word2Vec
```

```python
sentences = [
    "college life is fun and exciting",
    "students attend lectures daily",
    "exams are stressful for students",
    "friends make college life memorable",
    "assignments and projects are important",
    "library is a quiet place to study",
    "group study helps in understanding concepts",
    "college festivals are enjoyable",
    "students participate in sports and events",
    "hostel life teaches independence",
    "late night study is common during exams",
    "teachers guide students for better learning",
    "practical knowledge is important in college",
    "students prepare for placements",
    "internships help students gain experience"
]
```

```python
tokenized_sentence = [sentence.lower().split() for sentence in sentences]

model = Word2Vec(sentences=tokenized_sentence,vector_size=50,window=3,min_count=1,workers=4)
```

```python
word = "students"
print(f"Vector for '{word}':\n")
print(model.wv[word])
```

```python
print(f"\nTop 5 words similar to '{word}':\n")
similar_words = model.wv.most_similar(word,topn=5)
```

```python
for w, score in similar_words:
    print(f"{w} : {score}")
```
