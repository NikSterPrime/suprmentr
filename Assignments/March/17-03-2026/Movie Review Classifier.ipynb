```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Dataset (20+ sentences)
data = {
    "review": [
        "I love this movie",
        "This film was amazing",
        "Great acting and story",
        "Absolutely fantastic experience",
        "I enjoyed every moment",
        "It was a wonderful movie",
        "Brilliant performance by actors",
        
        "I hate this movie",
        "Worst film ever",
        "Terrible acting",
        "Not worth watching",
        "Very disappointing",
        "Awful storyline",
        "I regret watching this",
        
        "The movie was okay",
        "It was average",
        "Nothing special about it",
        "It was fine",
        "Just another movie",
        "Neither good nor bad"
    ],
    
    "sentiment": [
        "Positive","Positive","Positive","Positive","Positive","Positive","Positive",
        "Negative","Negative","Negative","Negative","Negative","Negative","Negative",
        "Neutral","Neutral","Neutral","Neutral","Neutral","Neutral"
    ]
}

df = pd.DataFrame(data)

# Step 2: Split data
X_train, X_test, y_train, y_test = train_test_split(
    df["review"], df["sentiment"], test_size=0.2, random_state=42
)

# Step 3: TF-IDF Vectorization (Bonus)
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Step 4: Train Model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Step 5: Predictions
y_pred = model.predict(X_test_vec)

# Step 6: Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Step 7: Test with custom input
test_sentences = [
    "This movie was awesome",
    "I did not like this film",
    "It was just okay"
]

test_vec = vectorizer.transform(test_sentences)
predictions = model.predict(test_vec)

print("\nCustom Predictions:")
for sentence, pred in zip(test_sentences, predictions):
    print(f"{sentence} --> {pred}")
```
