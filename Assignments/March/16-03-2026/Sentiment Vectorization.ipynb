```python
## Sentiment Vectorization
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
documents = [
    "I love this movie",
    "This movie is terrible",
    "Amazing acting",
    "Worst film ever"
]
```

```python
bow = CountVectorizer()
bow_matrix = bow.fit_transform(documents)
```

```python
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=bow.get_feature_names_out())

print("=== Bag of Words Matrix ===")
print(bow_df)
print("\n")
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())

print("=== TF-IDF Matrix ===")
print(tfidf_df)
print("\n")
```

```python
print("=== Comparison ===")
print("BoW counts word frequency (simple counts).")
print("TF-IDF gives importance to rare words and reduces weight of common words.\n")
print("Bag of Words DataFrame:\n", bow_df)
print("\nTF-IDF DataFrame:\n", tfidf_df)
```
