```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Step 1: Dataset (5 documents)
docs = [
    "Machine learning is used in artificial intelligence",
    "Artificial intelligence and deep learning are related fields",
    "Natural language processing is a part of AI",
    "Deep learning models require large data",
    "AI is transforming the world with smart systems"
]

# Step 2: TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(docs)

# Convert to DataFrame
df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

print("TF-IDF Matrix:\n", df)

# Step 3: Get top keywords per document
print("\nTop Keywords per Document:\n")
for i, row in df.iterrows():
    top_words = row.sort_values(ascending=False).head(3)
    print(f"Doc {i+1}: {list(top_words.index)}")
```
