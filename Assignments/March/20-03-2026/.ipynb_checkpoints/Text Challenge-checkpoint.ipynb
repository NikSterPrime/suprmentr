```python
import pandas as pd
import re

# Step 1: Dataset (20 messy sentences)
sentences = [
    "I luv this movie 😍",
    "This film is sooo goood!!!",
    "Worst movi ever 😡",
    "LOL that was funny 😂",
    "brb watching this again",
    "I cant belive this 😭",
    "Amaaaazing acting!!",
    "Not gud at all",
    "This is lit 🔥",
    "meh it was ok",
    "What a wast of time",
    "OMG I loved it 😍😍",
    "so boringgg",
    "gr8 movie",
    "idk what to say",
    "that ending tho 😳",
    "it was kinda coool",
    "bad scripttt",
    "nice one 👍",
    "wanna watch again??"
]

# Step 2: Slang dictionary
slang_dict = {
    "luv": "love",
    "LOL": "laughing",
    "brb": "be right back",
    "gud": "good",
    "gr8": "great",
    "idk": "I don't know",
    "OMG": "oh my god"
}

# Step 3: Function to detect emojis
def extract_emojis(text):
    return re.findall(r'[^\w\s,]', text)

# Step 4: Function to detect repeated characters (typos)
def detect_typos(text):
    return re.findall(r'(.)\1{2,}', text)

# Step 5: Analyze sentences
results = []

for sentence in sentences:
    words = sentence.split()
    
    slang_found = [word for word in words if word in slang_dict]
    emojis_found = extract_emojis(sentence)
    typos_found = detect_typos(sentence)
    
    results.append({
        "Sentence": sentence,
        "Slang": slang_found,
        "Emojis": emojis_found,
        "Typos": typos_found
    })

# Step 6: Convert to DataFrame
df = pd.DataFrame(results)

print("=== Analysis of Text Challenges ===\n")
print(df)

# Step 7: Basic Preprocessing Example
def preprocess(text):
    # Lowercase
    text = text.lower()
    
    # Replace slang
    for slang, full in slang_dict.items():
        text = text.replace(slang.lower(), full.lower())
    
    # Remove emojis and special characters
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    
    return text

print("\n=== Cleaned Sentences ===\n")
for s in sentences:
    print(preprocess(s))
```
