```python
import re
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Step 1: Sample messy sentences
sentences = [
    "This is AMAZING!!!",
    "I love this movie, it's awesome :)",
    "Worst film ever... so boring!!!",
    "Can you believe THIS??",
    "It was not good, not bad, just okay."
]

# Step 2: Define cleaning function
def clean_text(text):
    # Lowercase
    text = text.lower()
    
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove stopwords
    words = text.split()
    words = [word for word in words if word not in ENGLISH_STOP_WORDS]
    
    return " ".join(words)

# Step 3: Apply cleaning
cleaned = [clean_text(sentence) for sentence in sentences]

# Step 4: Create DataFrame for comparison
df = pd.DataFrame({
    "Original": sentences,
    "Cleaned": cleaned
})

print("=== Text Cleaning Result ===\n")
print(df)
```
