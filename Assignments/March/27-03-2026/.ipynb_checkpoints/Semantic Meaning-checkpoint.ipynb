```python
from gensim.models import Word2Vec

# Dataset
sentences = [
    ["king", "queen", "man", "woman"],
    ["boy", "girl", "man", "woman"],
    ["car", "bus", "train", "vehicle"],
    ["apple", "banana", "fruit"],
    ["dog", "cat", "animal"]
]

# Train model
model = Word2Vec(sentences, vector_size=50, min_count=1)

# Word pairs
pairs = [
    ("king", "queen"),
    ("car", "bus"),
    ("apple", "banana"),
    ("dog", "cat"),
    ("boy", "girl")
]

print("Semantic Similarity:\n")
for w1, w2 in pairs:
    similarity = model.wv.similarity(w1, w2)
    print(f"{w1} - {w2}: {similarity}")
```
