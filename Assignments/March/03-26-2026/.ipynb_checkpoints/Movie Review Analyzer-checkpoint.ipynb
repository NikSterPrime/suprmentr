```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# Training data
reviews = [
    "I love this movie",
    "This is amazing",
    "I hate this film",
    "Worst movie ever",
    "It is okay",
    "Not bad not great"
]

labels = ["Positive", "Positive", "Negative", "Negative", "Neutral", "Neutral"]

# Vectorization
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(reviews)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Test on 5 reviews
test_reviews = [
    "This movie is fantastic",
    "I did not like this",
    "It was average",
    "Best film ever",
    "Very boring movie"
]

X_test = vectorizer.transform(test_reviews)
predictions = model.predict(X_test)

print("Predictions:\n")
for r, p in zip(test_reviews, predictions):
    print(f"{r} → {p}")
```
